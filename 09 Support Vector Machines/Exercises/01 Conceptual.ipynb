__1.__ This problem involves hyperplanes in two dimensions.

(a)  Sketch the hyperplane $1 + 3X_1 - X_2 = 0$. Indicate the set of points for which $1 + 3X_1 - X_2 > 0$, as well as the set of points for which $1 + 3X_1 - X_2 < 0$.

```r
x <- seq(-1,3,by=.1)
hyper <- function(x) 1 + 3*x
y <- seq(hyper(-1),hyper(3),by=.1)
plot(x, hyper(x), type="l", xlab="X1", ylab="X2", xlim=c(-1,3), ylim=c(hyper(-1),hyper(3)))
X1 = c()
X2 = c()
for (x1 in x) {
    for (x2 in y) {
        X1 = c(X1, x1)
        X2 = c(X2, x2)
    }
}
color <- "red"
color[X2>hyper(X1)] <- "blue"
color[X2<hyper(X1)] <- "red"
points(X1, X2, col=color, cex=.1)
rm(list=ls())
```

__2.__ We have seen that in $p = 2$ dimensions, a linear decision boundary takes the form $\beta_0 + \beta_1 X_1 + \beta_2 X_2 = 0$. We now investigate a non-linear decision boundary.    
(a) Sketch the curve $(1+X_1)^2 + (2-X_2)^2 = 4$.
(b) On your sketch, indicate the set of points for which
\begin{align*}
(1+X_1)^2 + (2-X_2)^2 > 4\text{,}
\end{align*}
as well as the set of points for which
\begin{align*}
(1+X_1)^2 + (2-X_2)^2 \le 4\text{.}
\end{align*}
(c) Suppose that a classifier assigns an observation to the blue class if
\begin{align*}
(1+X_1)^2 + (2-X_2)^2 > 4\text{,}
\end{align*}
and to the red class otherwise. To what class is the observation $(0, 0)$ classified? $(-1, 1)$? $(2, 2)$? $(3, 8)$?  
(d) Argue that while the decision boundary in (c) is not linear in terms of $X_1$ and $X_2$, it is linear in terms of $X_1$, $X^2_1$, $X_2$, and $X^2_2$.  

```r
# Create a grid of values for X1 and X2
X1 <- seq(-5, 5, length.out = 100)
X2 <- seq(-5, 5, length.out = 100)

# Create a grid of (X1, X2) pairs
grid <- expand.grid(X1 = X1, X2 = X2)

# Calculate the corresponding Z values based on the equation
grid$Z <- (1 + grid$X1)^2 + (2 - grid$X2)^2 - 4

# Plot the contour of the curve
contour(
  x = unique(grid$X1),
  y = unique(grid$X2),
  z = matrix(grid$Z, nrow = length(unique(grid$X1)), ncol = length(unique(grid$X2))),
  levels = 0,
  xlab = "X1",
  ylab = "X2",
  main = expression((1 + X[1])^2 + (2 - X[2])^2 == 4)
)
points(grid$X1[grid$Z > 0], grid$X2[grid$Z > 0], col = "blue", pch = 16)
points(grid$X1[grid$Z <= 0], grid$X2[grid$Z <= 0], col = "red", pch = 16)
points(0, 0, col = "black", pch = 18)
points(-1, 1, col = "black", pch = 18)
points(2, 2, col = "black", pch = 18)
points(3, 8, col = "black", pch = 18)
```
