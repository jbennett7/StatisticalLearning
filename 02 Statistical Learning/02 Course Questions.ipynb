## 2.2 Dimensionality and Structured Models
### R.1
A hypercube with side length $1$ in d dimensions is defined to be the set of points ($x_1,x_2,\ldots,x_d)$ such that $0 \le x_i \le 1$ for all $j = 1,2,\ldots,d$. The boundary of the hypercube is defined to be the set of all points such that there exists a $j$ for which $0 \le x_j \le .05$ or $.95 \le x_j \le 1$ (namely, the boundary is the set of all points that have at least one dimension in the most extreme $10%$ of possible values). What proportion of the points in a hypercube of dimension $50$ are in the boundary? (hint: you may want to calculate the volume of the non-boundary region)

Please give your answer as a value between $0$ and $1$ with $3$ significant digits. If you think answer is $50.52%$, you should say $0.505$:

Answer: $0.995$

We know that the volume of the whole hypercube is 1^{50}=1. The volume of the interior of the hypercube is $0.9^{50}=0.005$. Thus, the volume of the boundary is $1-0.005=0.995$.

## 2.3 Model Selection and Bias-Variance Tradeoff
### R.1
True or False: A fitted model with more predictors will necessarily have a lower Training Set Error than a model with fewer predictors.

Answer: False
While we typically expect a model with more predictors to have lower Training Set Error, it is not necessarily the case. An extreme counterexample would be a case where you have a model with one predictor that is always equal to the response, compared to a model with many predictors that are random.

### R.2
While doing a homework assignment, you fit a Linear Model to your data set. You are thinking about changing the Linear Model to a Quadratic one. Which of the following is most likely true:

1. Using the Quadratic Model will decrease your Irreducible Error.
2. Using the Quadratic Model will decrease the Bias of your model.
3. Using the Quadratic Model will decrease the Variance of your model.
4. Using the Quadratic Model will decrease your Reducible Error.

Answer: 2 - Using the Quadratic Model will decrease the Bias of your model.

Introducing the quadratic term will make your model more complicated. More complicated models typically have lower bias at the cost of higher variance. This has an unclear effect on Reducible Error (could go up or down) and no effect on Irreducible Error.

## Chapter 2 Quiz
### Q.1
For each of the following parts, indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible model.  

The sample size $n$ is extremely large, and the predictors $p$ is small:  

Answer: Flexible is better  

A flexible model will allow us to take full advantage of our large sample size.

### Q.2
The number of predictors $p$ is extremely large, and the sample size $n$ is small:  

Answer: Flexible is worse  

The flexible model will cause overfitting due to the small sample size.  

### Q.3
The relationship between the predictors and response is highly non-linear:  

Answer: Flexible is better  

A flexible model will be necessary to find the nonlinear effect.  

### Q.4
The variance of the error terms, i.e. $\sigma^2=\text{Var}(\epsilon)$, is extremely high:

Answer: Flexible is worse  

A flexible model will cause us to fit too much of the noise in the problem.
