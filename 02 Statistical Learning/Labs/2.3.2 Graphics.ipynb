# 2.3.2 Graphics

The `plot()` function is the primary way to plot data in `R`.

```r
set.seed(1)
x <- rnorm(100)
y <- rnorm(100)
plot(x, y, xlab = "x-axis", ylab = "y-lab", main = "Plot")
```

We can also save the file as a `pdf` or `jpeg`

```r
pdf("Figure.pdf")
plot(x, y, col = "green")
dev.off()
```

```r
jpeg("Figure.jpeg")
plot(x, y, col = "skyblue")
dev.off()
```

The function `def.off()` indicates to `R` that we are done creating the plot. We can also copy and paste the image into a word document.

The `seq(a, b)` can create a sequence of data.

```r
seq(1,10)
x <- 1:10
```

```r
y <- x
f <- outer(x, y, function(x, y) cos(y) / (1 + x^2))
contour(x, y, f)
contour(x, y, f, nlevels = 45, add = T)
fa <- (f - t(f)) / 2
contour(x, y, fa, nlevels = 15)
```

```r
image(x, y, fa)
persp(x, y, fa)
persp(x, y, fa, theta = 30)
persp(x, y, fa, theta = 30, phi = 20)
persp(x, y, fa, theta = 30, phi = 70)
persp(x, y, fa, theta = 30, phi = 40)
```
