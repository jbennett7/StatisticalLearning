## 4.3 Multivariate Logistic Regression
### R.1
Suppose we collect data for a group of students in a statistics class with variables $X_1=$ hours studied, $X_2=$ undergrad GPA, and $Y=$ receive an A. We fit a logistic regression and produce estimated coefficients $\hat{\beta_0} = -6, \hat{\beta_1}=0.05, \hat{\beta_2}=1$.  

Estimate the probability that a student who studies for 40h and has an undergrad GPA of 3.5 gets an A in the class (within 0.01 accuracy):  

__SOLUTION__  
\begin{align}
log{\left( \frac{p(X)}{1 - p(X)} \right)} &= \beta_0 + \beta_1 X_1 + \beta_2 X_2 \\
p(X) &= \frac{ e^{\beta_0 + \beta_1 X_1 + \beta_2 X_2} } { 1 + e^{\beta_0 + \beta_1 X_1 + \beta_2 X_2}} \\
\\
\\
p(X) &= \frac{0.6065307}{1.606531} = 0.38
\end{align}

## 4.4 Logistic Regression - Case-Control Sampling and Multiclass
### R.1
In which of the following problems is Case/Control Sapling LEAST likely to make a positive impact?  

1. Predicting a shopper's gender based on the products they buy.
2. Finding predictors for a certain type of cancer.
3. Predicting if an email is Spam or Not Spam.

## 4.5 Discriminant Analysis
### R.1
Suppose that in Ad Clicks (a problem where you try to model if a user will click on a particular ad) it is well known that the majority of the time an ad is shown it will not be clicked. What is another way of saying that?  
1. Ad Clicks have a low Prior Probability
2. Ad Clicks have a high Prior Probability
3. Ad Clicks have a low Density

Answer 1, Ad Clicks have a low Prior Probability  

Whether or not an ad gets clicked is a Qualitative Variable. Thus, it does not have a density. The Prior Probability of Ad Clicks is low because most ads are not clicked.

## 4.6 Gaussian Discriminant Analysis - One Variable
### R.1
Which of the following is NOT a linear function in $x$:  

1. $f(x) = a + b^2x$
2. The discriminant function from LDA
3. $\delta_k(X) = x\frac{\mu_k}{\sigma^2} - \frac{\mu^2_k}{2\sigma^2} + \log{(\pi_k)}$
4. $\text{logit}(P(y=1|x))$ where $P(y=1|x)$ is as in logistic regression
5. $P(y=1|x)$ from logistic regression

Answer: 5, $P(y=1|x)$ from logistic regression  

$P(y=1|x)$ from logistic regression is not linear because it involves both an exponential function of $x$ and a ratio. Notice that $f(x)=a+b^2x$ is not a linear function of $b$, but is a linear function of $x$.

## 4.7 Gaussian Discriminant Analysis - Many Variables
### R.1
Why does Total Error keep going down on the graph on page 34 of the notes, even though the False Negative Rate increases?  
1. The False Negative Rate does not affect Total Error.
2. A higher False Negative Rate generally decreases Total Error.
3. Positive responses are so uncommon that their impact on the Total Error is small.
4. All of the above.

Answer: 3, Positive responses are so uncommon that their impact on the Total Error is small.  

The Total Error is a weighted average of the False Positive Rate and False Negative Rate. The weights are determined by the Prior Probabilities of Positive and Negative Responses. Although the False Negative Rate might be high, the prior for positive responses is very small.

## 4.8 Generalized Linear Models
Which of the following statements is true for linear logistic regression?  
1. The probability of a binary outcome is represented as a linear function of the input features.
2. The model specifies a form for the mean as well as the variance of the response.
3. The model forces the modeled mean to lie between zero and one.
4. The model is fit by weighted least squares.

Answer: 2, 3  

The logit of the mean is linear in the features.  
If the mean is $\mu$ then the variance is given by $\mu(1-\mu)$.  
The inverse logit transform is $e^{\eta}/(1+e^{\eta})$ which lies between 0 and 1 by construction.  
The model is fit by maximum-likelihood, which is often implemented as "iteratively-reqeighted least squares".

## 4.9 Quadratic Discriminant Analysis and Naive Bayes
### R.1
Which of the following statements best explains the relationship between Quadratic Discriminant Analysis and naive Bayes with Gaussian distributions in each class?  
1. Quadratic Discriminant Analysis is a more flexible class of models than naimve Bayes
2. Quadratic Discriminant Analysis is a less flexible class of models than naive Bayes
3. Quadratic Discriminant Analysis is an equivalently flexible class of models to naive Bayes
4. For some problems Quadratic Discriminant Analysis is more flexible than naive Bayes, for others the opposite is true.

Answer: 1  

With Gaussian distributions, naive Bayes is equivalent to Quadratic Discriminant Analysis with the additional requirement that each class covariance matrix $\sum_k$ be diagonal. Thus, Quadratic Discriminant Analysis is more flexible.
