__1.__ We perform best subset, forward stepwise, and backward stepwise selection on a single data set. For each approach, we obtain $p + 1$ models, containing $0, 1, 2,\ldots, p$ predictors. Explain your answers:  

&nbsp;&nbsp;(a) Which of the three models with $k$ predictors has the smallest _training_ RSS?  
&nbsp;&nbsp;(b) Which of the three models with $k$ predictors has the smallest _test_ RSS?  
&nbsp;&nbsp;(c) True or False:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;i.&nbsp;&nbsp;&nbsp;&nbsp;The predictors in the _k_-variable model identified by forward stepwise are a subset of the predictors in the $(k+1)$-variable model identified by forward stepwise selection.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;ii.&nbsp;&nbsp;&nbsp;The predictors in the _k_-variable model identified by backward stepwise are a subset of the predictors in the $(k + 1)$-variable model identified by backward stepwise selection.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;iii.&nbsp;&nbsp;The predictors in the _k_-variable model identified by backward stepwise are a subset of the predictors in the $(k + 1)$-variable model identified by forward stepwise selection.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;iv.&nbsp;&nbsp;&nbsp;The predictors in the _k_-variable model identified by forward stepwise are a subset of the predictors in the $(k + 1)$-variable model identified by backward stepwise selection.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;v.&nbsp;&nbsp;&nbsp;&nbsp;The predictors in the _k_-variable model identified by best subset are a subset of the predictors in the $(k + 1)$-variable model identified by best subset selection.

__SOLUTION__  

&nbsp;&nbsp;(a) Best subset selection will have the smallest _training_ RSS because all models will be considered and the model that involves all of the variables will have the lowest RSS.  
&nbsp;&nbsp;(b) The model that has the best _training_ RSS is the model we are looking for. It is hard to say which method will produce the best model, but the best subset has the best chance.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;However, stepwise forward or backward may choose the best model faster.  
&nbsp;&nbsp;(c)  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;i.&nbsp;&nbsp;&nbsp;&nbsp;__True__, forward stepwise selection is done by adding variables to the model.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;ii.&nbsp;&nbsp;&nbsp;__True__, backward stepwise selection is done by removing variables from the model.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;iii.&nbsp;&nbsp;__False__, there is no relation between the models chosen by forward stepwise selection and backward stepwise selection.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;iv.&nbsp;&nbsp;&nbsp;__False__, again no relation.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;v.&nbsp;&nbsp;&nbsp;&nbsp;__False__, The model with $(k+1)$ predictors is obtained by selecting among all possible models with $(k+1)$ predictors,  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;and so does not necessarily contain all the predictors selected for the _k_-variable model.

\lambda \sum^p_{j=1} \beta^2_j
\end{align}
for a particular value of $\lambda$. For parts (a) throug (e), indicate which of i. through v. is correct. Jsutify your answer.  

&nbsp;&nbsp;(a) As we increase $\lambda$ from $0$, the training RSS will:  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;i.&nbsp;&nbsp;&nbsp;&nbsp;Increase initially, and then eventually start decreasing in an inverted U shape.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;ii.&nbsp;&nbsp;&nbsp;Decrease initially, and then eventually start increasing in a U shape.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;iii.&nbsp;&nbsp;Steadily increase.  
&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;iv.&nbsp;&nbsp;&nbsp;Steadily decrease.

&nbsp;&nbsp;(b) The test RSS will initially decrease but will eventually increase in a U shape.  
&nbsp;&nbsp;(c) The variance will steadily decrease.  
&nbsp;&nbsp;(d) The (squared) bias will steadily increase.  
&nbsp;&nbsp;(e) The irreducible error is independent of the model and thus will remain constant.

__5.__ It is well-known that ridge regression tends to give similar coefficient values to correlated variables, whereas the lasso may give quite different coefficient values to correlated variables. We will now explore this property in a very simple setting.  

Suppose that $n=2$, $x_{11} = x_{12}$, $x_{21} = X_{22}$. FUrthermore suppose that $y_1 + y_2 = 0$ and $x_{11} + x_{21} = 0$ and $x_{12} + x_{22} = 0$, so that the estimate for the intercept in a least squares, ridge regression, or lasso model is zero: $\hat{\beta_0} = 0$.  

&nbsp;&nbsp;(a) Write out the ridge regression optimization problem in this setting.  
&nbsp;&nbsp;(b) Argue that in this setting, the ridge coefficient estimates satisfy $\hat{\beta_1} = \hat{\beta_2}$.  
&nbsp;&nbsp;(c) Write out the lasso optimization problem in this setting.  
&nbsp;&nbsp;(d) Argue that in this setting, the lasso coefficients $\hat{\beta_1}$ and $\hat{\beta_2}$ are not unique&mdash;in other words, there are many possible solutions to the optimization problem in (c). Describe these solutions.  

__SOLUTION__  

&nbsp;&nbsp;(a) For this setting, $x_{11} = x_{12} = x_1$ and $x_{21} = x_{22} = x_2$, the ridge regression model seeks to minimize:
\begin{align}
(y_1 - \hat{\beta_1}x_1 - \hat{\beta_2}x_1)^2 +
(y_2 - \hat{\beta_1}x_2 - \hat{\beta_2}x_2)^2 +
\lambda (\hat{\beta_1}^2 + \hat{\beta_2}^2)
\end{align}
&nbsp;&nbsp;(b) By taking the derivatives of the above expression with respect to $\hat{\beta_1}$ and $\hat{\beta_2}$ and setting them equal to $0$, we obtain respectively:
\begin{align}
\hat{\beta_1}(x^2_1 + x^2_2 + \lambda) +
\hat{\beta_2}(x^2_1 + x^2_2)           =
y_1x_1 + y_2 x_2
\end{align}
and
\begin{align}
\hat{\beta_1}(x^2_1 + x^2_2)           +
\hat{\beta_2}(x^2_1 + x^2_2 + \lambda) =
y_1x_1 + y_2 x_2
\end{align}
By subtracting the two expressions above we get $\hat{\beta_1} = \hat{\beta_2}$.  
&nbsp;&nbsp;(c) For this setting, $(x_{11} = x_{12} = x_1$ and $x_{21} = x_{22} = x_2)$, the lasso regression model seeks to minimize:
\begin{align}
(y_1 - \hat{\beta_1}x_1 - \hat{\beta_2}x_1)^2 +
(y_2 - \hat{\beta_1}x_2 - \hat{\beta_2}x_2)^2 +
\lambda (|\hat{\beta_1}| + |\hat{\beta_2}|)
\end{align}
&nbsp;&nbsp;(d) We use this form of the lasso optimization problem:
\begin{align}
(y_1 - \hat{\beta_1}x_1 - \hat{\beta_2}x_1)^2 +
(y_2 - \hat{\beta_1}x_2 - \hat{\beta_2}x_2)^2 \ \ \ \ \ \ 
\text{ subject to }\ \ \ \ 
|\hat{\beta_1}| + |\hat{\beta_2}| \le s
\end{align}
Geometrically the lasso constraint takes the form of a diamond centered at the origin of the plane $\left(\hat{\beta_1}, \hat{\beta_2}\right)$ which intersects the axes at the distance $s$ from the origin. By using the setting of this problem $(x_{11} = x_{12} = x_1, x_{21} = x_{22} = x_2, x_1 + x_2 = 0$ and $y_1 + y_2 = 0)$, we have to minimize the expression
\begin{align}
2 \left[ y_1 - \left(\hat{\beta_1} + 
         \hat{\beta_2}\right)x_1 \right]^2 \gt 0
\end{align}
This optimization problem has a simple solution: $\hat{\beta_1} + \hat{\beta_2} = y_1 / x_1$. Geometrically, this is a line parallel to the edge of the diamon of the constraints. Now, solutions to the lasso optimization problem are contours of the function $\left[ y_1 - \left( \hat{\beta_1} + \hat{\beta_2}\right)x_1 \right]^2$ that intersects the diamond of the constraints. So, the entire edge $\hat{\beta_1} + \hat{\beta_2} = s$ (as is the edge $\hat{\beta_1} + \hat{\beta_2} = -s$) is a potential solution to the lasso optimization problem. Thus, the lasso optimization problem has a whole set of solutions instead of a unique one:
\begin{align}
\left\{
  \left(\hat{\beta_1}, \hat{\beta_2}\right):
    \hat{\beta_1} + \hat{\beta_2} = s
    \text{ with }
    \hat{\beta_1}, \hat{\beta_2} \ge 0
    \text{ and }
    \hat{\beta_1} + \hat{\beta_2} = -s
    \text{ with }
    \hat{\beta_1}, \hat{\beta_2} \le 0
\right\}
\end{align}

```r
y <- 3
lambda <- 2
beta <- seq(-10, 10, 0.1)
plot(beta, (y - beta)^2 + lambda * beta^2, pch = 20, xlab = "beta", ylab = "Ridge optimization")
beta.est <- y / (1 + lambda)
points(beta.est, (y - beta.est)^2 + lambda * beta.est^2, col = "red", pch = 4, lwd = 5)
```

We may see that the function is minimized at $\beta = y / (1 + \lambda)$.

```r
y <- 3
lambda <- 2
beta <- seq(-10, 10, 0.1)
plot(beta, (y - beta)^2 + lambda * abs(beta), pch = 20, xlab = "beta", ylab = "Ridge optimization")
beta.est <- y - lambda / 2
points(beta.est, (y - beta.est)^2 + lambda * abs(beta.est), col = "red", pch = 4, lwd = 5)
```

We may see that the function is minimized at $\beta = y - \lambda / 2$ as $y > \lambda/2$.
