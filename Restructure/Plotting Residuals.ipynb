# Plotting Residuals

```r
plot.residuals <- function(x, fit, col="red") {
    for(idx in 1:length(fit$fit)){
        lines(c(x[idx],x[idx]),c(fit$fit[idx],fit$fit[idx]+fit$residual[idx]),col=col)
    }
}
```

```r
set.seed(1)
n <- 45
xlim <- c((-pi+1.3),(pi-.3))
x <- seq(xlim[1],xlim[2], length.out=n)
y <- jitter(sin(x),amount=.4)
y.known <- sin(x)
plot(x,y, pch=18)
fit <- lm(y ~ x)
lines(x,fit$fit,col="blue")
plot.residuals(x,fit)
```
